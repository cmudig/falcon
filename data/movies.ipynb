```python
import pandas as pd
import pyarrow as pa
```

```python
df = pd.read_json('./movies.json')
df = df[df.columns[1:]]
```

```python
df
```

```python
def pandas_to_arrow_file(df, filename="default.arrow"):
	table = pa.Table.from_pandas(df)
	writer = pa.RecordBatchFileWriter(filename, table.schema)
	writer.write(table)
	writer.close()
```

```python
pandas_to_arrow_file(df, "movies-3k.arrow")
```
