```python
import pandas as pd
import pyarrow as pa
import dataio as io
```

```python
df = pd.read_json('./movies-3k.json')
df.columns = [i.replace(' ','_') for i in df.columns]
```

```python
df.to_json("movies-3k.json", orient="records")
```

```python
df.head()
```

```python
# encode the data column correctly
df["Release_Date"] = pd.to_datetime(df["Release_Date"])
```

```python
df["Release_Date"] = df["Release_Date"].dt.to_pydatetime()
```

```python
df.head()
```

```python
df.loc[9, "Release_Date"]
```

```python
table = pa.Table.from_pandas(df)
```

```python

io.table_to_arrow_file(table, "movies-3k.arrow")
io.table_to_parquet_file(table, "movies-3k.parquet")
```
