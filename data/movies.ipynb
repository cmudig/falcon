```python
import pandas as pd
import pyarrow as pa
import dataio as io
```

```python
df = pd.read_json('./movies.json')
df = df[df.columns[1:]] # remove the first column title
```

```python
df.head()
```

```python
table = pa.Table.from_pandas(df)
io.table_to_arrow_file(table, "movies-3k.arrow")
io.table_to_parquet_file(table, "movies-3k.parquet")
```
