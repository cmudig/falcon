```python
import pyarrow as pa
import dataio
import pandas as pd
```

```python
df = pd.read_json('movies.json', encoding='utf-8')
```

```python
# convert Title to string dtype
df["Title"] = df["Title"].astype(str)
df
```

```python
table = pa.Table.from_pandas(df)
```

```python
dataio.table_to_arrow_file(table, 'exp1.arrow')
dataio.table_to_parquet_file(table, 'exp1.parquet')
```

```python
def filter_between(table, column_name, a, b):
    return table[(table[column_name] >= a) & (table[column_name] <= b)]
```

```python
filter_between(df, "IMDB_Rating", 6.41, 8.52)
```
